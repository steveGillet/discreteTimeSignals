```python
import numpy as np
import matplotlib.pyplot as plt

n = np.array((-3,-2,-1,0,1,2,3))
x = np.array((-1,2,1,-2,0,0,0))
y = np.array((0,1,2,1,0,1,0))
def plot(x,n,ylabel):
  plt.stem(n, x)
  plt.xticks(n)
  plt.xlabel("n")
  plt.ylabel(ylabel)
  plt.show()
plot(x,n,'x[n]')
plot(y,n,'y[n]')
```

```python
w1 = np.flip(x) + y
print(w1)
plot(w1,n,"w1[n] (x[-n]+y[n])")
```

```python
y2 = np.compress((n % 2) == 0,y)
n2 = np.compress((n % 2) == 0,n)
y2Fill = np.zeros((7))
for i in range(len(n2)):
  for j in range(len(y2Fill)):
    if(j==n2[i]+3):
      y2Fill[j] = y2[i]

w2 = x - y2Fill
print(w2)
plot(w2,n,"w2[n] (x[n]-y[2n])")
```

```python
x3 = np.compress((n % 2) == 0,x)
x3Fill = np.zeros((7))
y3 = y2
n3 = n2
y3Fill = y2Fill
for i in range(len(n3)):
  for j in range(len(x3Fill)):
    if(j==n3[i]+3):
      x3Fill[j] = x3[i]

w3 = x3Fill + y3Fill
print(w3)
plot(w3,n,"w3[n] (x[2n]+y[2n])")
```

```python
def decimation(seq, ind, indShift, D):
  seqFill = np.zeros((len(seq)))
  seq = np.compress((ind % D) == 0,seq)
  ind = np.compress((ind % D) == 0,ind)
  for i in range(len(ind)):
    for j in range(len(seqFill)):
      if(j==ind[i]+indShift):
        seqFill[j] = seq[i]
  return seqFill

x4 = decimation(x,n,3,3)
w4 = -x4*y
print(w4)
plot(w4,n,"w4[n] (x[-3n]y[n])")
```
